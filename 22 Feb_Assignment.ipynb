Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

A DataFrame in pandas is a two-dimensional labeled data structure that represents a table of data with rows and columns. It is a fundamental object in the pandas library and is designed for efficient data manipulation and analysis.

A DataFrame is a collection of Pandas Series objects, where each Series represents a column of data. While a Series is a one-dimensional labeled array, a DataFrame provides a tabular structure with both row and column labels. This allows for more flexible data manipulation and analysis, as it can handle multiple columns and rows of data simultaneously.

Here's an example to illustrate the difference between a DataFrame and a Series:

```python
#DataFrame
Data={"Name":["Harshit","Arpit","Abhilasha","Sagar","Sadhana","Gotu","Lucky","Megha","Bhuvi","Khushi","Ramesh"],"Age":[27,33,30,62,57,4,32,31,2,0.6,65],"Gender":["M","M","F","M","F","F","M","F","F","F","M"],"Married Status":["Unmarried","Married","Married","Married","Married","Unmarried","Married","Married","Unmarrire","Unmarried","Married"]}
df2=pd.DataFrame(Data)
df2
```

```python
Age=[27,33,30,62,57,4]
s2=pd.Series(Age)
s2
```

Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

Pandas provides a wide range of functions to manipulate data in a DataFrame. Here are some common functions along with examples of when you might use them:

```python
#DataFrame
Data={"Name":["Harshit","Arpit","Abhilasha","Sagar","Sadhana","Gotu","Lucky","Megha","Bhuvi","Khushi","Ramesh"],"Age":[27,33,30,62,57,4,32,31,2,0.6,65],"Gender":["M","M","F","M","F","F","M","F","F","F","M"],"Married Status":["Unmarried","Married","Married","Married","Married","Unmarried","Married","Married","Unmarrire","Unmarried","Married"]}
df2=pd.DataFrame(Data)
df2

#head and tail
df2_head=df2.head()
df2_tail=df2.tail()
#first 5 data from df2
df2_head

#last 5 data from df2
df2_tail

#info: for information about the df2
df2_info=df2.info()
df2_info

#describe: describe all the static data related to df2
df2_describe=df2.describe()
df2_describe

#sortvalue: sorting the data of any column by using sortvalue
df2_sort=df2.sort_values('Age')
df2_sort

#sort in descending
df2_des=df2.sort_values('Age',ascending=False)
df2_des

#groupby(): This function is used for grouping rows based on one or more columns. It allows you to perform aggregations, calculations, or apply functions on subsets of the data based on the groups.
df2_group=df2.groupby('Name')['Age'].mean()
df2_group

#drop(): This function is used to drop specified rows or columns from the DataFrame. It is useful for removing unnecessary or redundant data.
df2_drop=df2.drop('Age',axis=1)
df2_drop
```

These are just a few examples of the functions available in Pandas to manipulate data in a DataFrame. Pandas offers many more powerful functions for data selection, filtering, merging, reshaping, and more, allowing you to perform complex data operations with ease.

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

A mutable object can be modified after it is created, allowing for changes to its values, addition of new elements, or removal of existing elements. Let's clarify the mutability of each object:

Series: A Series is mutable, meaning you can modify its values. You can change specific values in a Series or add new values to it.

DataFrame: A DataFrame is mutable, allowing you to modify its values, add or remove columns, and update rows. You can change values in specific cells, add or delete columns, or modify existing rows.

Panel: Panels were a data structure in pandas that used to handle three-dimensional data. However, as of pandas version 0.20.0, the Panel object is deprecated, and it is recommended to use multi-index DataFrames or other data structures instead. Panels were not mutable.

To summarize, both Series and DataFrame in pandas are mutable, meaning you can modify their contents. However, Panel, which is deprecated, is not mutable.
