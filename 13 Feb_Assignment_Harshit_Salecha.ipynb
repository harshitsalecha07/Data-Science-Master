```python
'''Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.'''
```

###### When creating a custom exception in Python, it is recommended to inherit from the Exception class, which is the base class for all exceptions in Python.

###### The reason for this is that the Exception class already contains a lot of the necessary functionality for creating and handling exceptions. For example, it has a constructor that takes an optional error message and stores it in an instance variable, making it easy to create custom exception classes with informative error messages. Additionally, it has methods like __str__() that allow you to define how the exception should be represented as a string.

###### By inheriting from Exception, your custom exception class will also be able to take advantage of existing exception handling mechanisms in Python. For example, if you raise your custom exception in a try block, it can be caught and handled by an appropriate except block further up the call stack.

```python
### Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

###### ArithmeticError is an error that will occur during any arithmetic failure found in by the console the run-time of the Python program. In python there is so many types of Arithmetic Errors.
###### 1.) ZeroDivisionError
###### 2.) OverflowError
###### 3.) UnderflowError
###### 4.) FloatingPointError
###### 5.) ValueError

```python
#Example of ZeroDivisionError
try:
    a=10
    b=0
    print(a/b)    ### In this step we are dividing non-zero number by zero thats why system generate a Arithmetic error i.e ZeroDivsionError
except ZeroDivisionError as e:
    print(e)
```

```python
#Example of ValueError
try:
    a=int("Harshit")   ### In this step we are going to convert alphabatical character into integer thats why system generate a value error
    print(a)
except ValueError as e:
    print(e)
```

```python
### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

### LookupError:
###### Lookup error is an error that occurs only then the console will not able to find the respective key or index in the input. 
### Types of LookupError
###### 1.) IndexError: IndexError occurs only then the given index is not found.
###### 2.) KeyError: KeyError occurs only then the given Key is not found.

```python
### Example of Index 
try:
    a=[1,2,3]
    a[5]     ### In this step console is not able to find the 5th index from the given list so it will generate IndexError
except IndexError as e:
    print(e)
```

```python
### Example of KeyError
try: 
    dic1={"a":1,"b":2,"d":3}
    dic1["c"]
except KeyError as e:
    print("Key {} is not found".format(e))
```

```python
### Q5. Explain ImportError. What is ModuleNotFoundError?
```

###### ImportError is an error that occurs only then the imported module is not found

```python
try:
    import xyz
    print(xyz) 
except ModuleNotFoundError as e:
    print(e)
```
