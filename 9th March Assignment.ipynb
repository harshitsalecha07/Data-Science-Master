```python
Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

Probability Mass Function (PMF):
The Probability Mass Function (PMF) is used for discrete random variables. It gives the probability that a discrete random variable takes on a specific value. In other words, it provides the probability of each possible outcome in the sample space.

Example of PMF:
Let's consider rolling a fair six-sided die. The PMF for this scenario would be:
- P(X = 1) = 1/6
- P(X = 2) = 1/6
- P(X = 3) = 1/6
- P(X = 4) = 1/6
- P(X = 5) = 1/6
- P(X = 6) = 1/6

Probability Density Function (PDF):
The Probability Density Function (PDF) is used for continuous random variables. Unlike PMF, which deals with individual probabilities, the PDF represents the likelihood of a continuous random variable falling within a specific range of values. It gives the relative likelihood of observing different values, but the probability of getting any single precise value is technically zero.

Example of PDF:
A common example of a continuous PDF is the normal distribution. The PDF of the standard normal distribution is given by the formula:
f(x) = (1 / √(2π)) * e^(-x^2/2)

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Cumulative Density Function (CDF):
The Cumulative Density Function (CDF) is used for both discrete and continuous random variables. For a given value 'x', the CDF gives the probability that the random variable is less than or equal to 'x'. In other words, it provides the cumulative probability up to a specific point.

Example of CDF:
For the fair six-sided die example, the CDF would be:
- F(X ≤ 1) = 1/6
- F(X ≤ 2) = 2/6 = 1/3
- F(X ≤ 3) = 3/6 = 1/2
- F(X ≤ 4) = 4/6 = 2/3
- F(X ≤ 5) = 5/6
- F(X ≤ 6) = 1

Why CDF is used?
The CDF is useful because it provides a complete picture of the probability distribution, giving the cumulative probabilities up to each point. It helps in calculating the probabilities of events falling within certain ranges or intervals, which is valuable for statistical analysis and making decisions based on probabilities.

Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

The normal distribution is widely used as a model in various real-world situations, especially when dealing with natural phenomena and large samples. Some examples include:
1. Heights of a population: The heights of people in a population often follow a normal distribution.
2. Test scores: Scores on standardized tests like IQ tests or SATs are often modeled using the normal distribution.
3. Errors in measurements: Errors in measurements or experimental data often follow a normal distribution.
4. Natural phenomena: Many natural phenomena, like the distribution of rainfall amounts or the weight of seeds from a plant, can be approximated by a normal distribution.

The normal distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ). The mean determines the central tendency of the distribution, representing the peak or center point. The standard deviation controls the spread or dispersion of the data points around the mean. A larger standard deviation leads to a broader and flatter distribution, while a smaller standard deviation results in a narrower and taller distribution.

Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal Distribution.

The importance of the normal distribution lies in its prevalence in nature and its properties, which make it a fundamental tool in statistics and probability theory. Some key reasons for its importance are:
1. Central Limit Theorem: The normal distribution is central to the Central Limit Theorem, which states that the sampling distribution of the sample means of large enough samples from any distribution will be approximately normally distributed. This allows statisticians to make inferences about population parameters based on sample means.
2. Statistical Inference: Many statistical tests and methods rely on the assumption of normality in the data. The normal distribution provides a reference point for making statistical inferences and hypothesis testing.
3. Probability Calculations: The normal distribution's well-defined properties and mathematical simplicity make it easier to calculate probabilities and percentiles for specific events.
4. Data Modeling: In many cases, real-life data can be approximated by a normal distribution, allowing for the use of normal-based models for analysis and predictions.

Real-life examples of normal distribution:
1. IQ Scores: Intelligence quotient (IQ) scores are often normally distributed, with most people clustering around the average IQ score.
2. Heights and Weights: In a large population, heights and weights of individuals often follow a normal distribution.
3. Exam Scores: The scores of a well-designed exam in a large student population can approximate a normal distribution.
4. Errors in Measurements: Errors in measurements from various instruments often follow a normal distribution.

Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

Bernoulli Distribution:
The Bernoulli distribution is a discrete probability distribution used for a single binary (two possible outcomes) random variable. It represents the outcome of a single Bernoulli trial, where the event can be either a "success" (usually denoted as 1) with probability 'p' or a "failure" (usually denoted as 0) with probability '1 - p'. The distribution is characterized by the parameter 'p', the probability of success.

Example of Bernoulli Distribution:
Tossing a fair coin can be modeled using a Bernoulli distribution. Let's say we define "heads" as a success (1) and "tails" as a failure (0). The probability of getting heads (success) is 0.5 (p = 0.5) in this case.

Difference between Bernoulli Distribution and Binomial Distribution:
1. Number of Trials:
   - Bernoulli Distribution: Represents a single trial or event with only two possible outcomes (success or failure).
   - Binomial Distribution: Represents the number of successes in a fixed number of independent Bernoulli trials.

2. Random Variable:
   - Bernoulli Distribution: Has a single binary random variable (0 or 1).
   - Binomial Distribution: Has a discrete random variable representing the number of successes (0, 1, 2, ..., n) in 'n' trials.

3. Parameters:
   - Bernoulli Distribution: Characterized by a single parameter 'p', the probability of success in a single trial.
   - Binomial Distribution: Characterized by two parameters: 'n', the number of trials, and 'p', the probability of success in each trial.

Q6: Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 

60? Use the appropriate formula and show your calculations.

To find the probability that a randomly selected observation will be greater than 60, we need to use the Z-score formula and then find the corresponding probability from the standard normal distribution table.

The Z-score formula is: Z = (X - μ) / σ

Where:
X = Value of interest (in this case, 60)
μ = Mean of the dataset (given as 50)
σ = Standard deviation of the dataset (given as 10)

Now, calculate the Z-score:

Z = (60 - 50) / 10
Z = 1

Next, we find the probability of Z being greater than 1 from the standard normal distribution table or using a statistical software/tool. For a Z-score of 1, the corresponding probability is approximately 0.8413.

So, the probability that a randomly selected observation will be greater than 60 is approximately 0.8413 or 84.13%.

Q7: Explain uniform Distribution with an example.

Uniform Distribution:
The uniform distribution is a continuous probability distribution where every value within a given interval has an equal probability of occurring. In other words, all outcomes in the range are equally likely to happen.

Example of Uniform Distribution:
An example of a uniform distribution is rolling a fair six-sided die. In this case, each face of the die has an equal chance of landing face-up. The probability of rolling any particular number (1, 2, 3, 4, 5, or 6) is 1/6, making it a uniform distribution.

Q8: What is the z score? State the importance of the z score.

Z-score (also known as standard score):
The z-score is a measure of how many standard deviations a data point is from the mean. It is used to standardize data in a way that allows for comparison between different datasets with different scales or units.

Z-score formula: Z = (X - μ) / σ

Where:
X = Data point
μ = Mean of the dataset
σ = Standard deviation of the dataset

Importance of the z score:
1. Standardization: The z-score transforms data into a standardized scale, centered around the mean and measured in units of standard deviation. This allows for meaningful comparisons and assessments across different datasets.
2. Outlier Detection: Z-scores help identify outliers, as data points with high or low z-scores are considered extreme relative to the dataset.
3. Probability Calculations: Z-scores are used to find probabilities and percentiles in the standard normal distribution table.
4. Hypothesis Testing: Z-scores play a crucial role in hypothesis testing and determining the statistical significance of results.

Q9: What is the Central Limit Theorem? State the significance of the Central Limit Theorem.

Central Limit Theorem:
The Central Limit Theorem (CLT) is a fundamental concept in statistics. It states that when independent random variables are added, their sum tends to follow a normal distribution, regardless of the shape of the original variables' distributions. Additionally, as the sample size increases, the distribution of the sample mean approaches a normal distribution, even if the original population is not normally distributed.

Significance of the Central Limit Theorem:
1. Allows Approximation: The CLT enables the use of the normal distribution to approximate the distribution of sample means, even when the population distribution is unknown or not normal.
2. Foundation of Statistical Inference: The CLT forms the basis for many statistical tests and confidence interval calculations that rely on the assumption of normality.
3. Sample Size Impact: The CLT shows that as sample size increases, the distribution of the sample mean becomes more normal, making larger samples more reliable for making inferences about the population.
4. Widely Applicable: The CLT is applicable to a wide range of situations, making it one of the most important and frequently used theorems in statistics.

Q10: State the assumptions of the Central Limit Theorem.

The Central Limit Theorem (CLT) is powerful but relies on certain assumptions to hold true:
1. Random Sampling: The sample should be selected randomly from the population of interest.
2. Independence: Each data point in the sample must be independent of others, meaning that the value of one observation should not influence the value of another.
3. Sufficient Sample Size: The sample size should be sufficiently large. A common rule of thumb is that the sample size should be at least 30 for the CLT to work reasonably well.
4. Finite Variance: The population from which the samples are drawn must have a finite variance. If the population variance is infinite, the CLT may not apply.
5. Sample Mean Calculation: The theorem applies to the distribution of the sample mean, regardless of the original distribution of the population.

If these assumptions are met, the Central Limit Theorem allows us to use the normal distribution to make inferences about the population, even when the population's distribution is not known or is not normal.
```
