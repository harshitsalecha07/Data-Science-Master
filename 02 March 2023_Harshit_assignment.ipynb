Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data

plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot, also known as a scatter diagram or scatter graph, is a type of data visualization that displays the relationship between two numerical variables. 

It is commonly used to identify the correlation or pattern between two variables and to assess the strength and direction of their relationship.

In a scatter plot, each data point is represented by a dot or marker on a two-dimensional Cartesian coordinate system. 

The x-axis represents one variable, and the y-axis represents the other variable. The position of each data point on the plot corresponds to the values of the variables it represents.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This is the x axis data")
plt.ylabel("This is the y axis data")
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot, also known as a bar chart or bar graph, is a type of data visualization that displays categorical data using rectangular bars. 

It is used to compare different categories or groups and visualize their relative sizes or frequencies.

Bar plots are particularly useful when you want to represent discrete data or compare data across different categories. 

They provide a visual representation of the data distribution, making it easy to identify patterns, trends, and variations between the categories.

To create a bar plot, you typically have categorical data on one axis (usually the x-axis) and numerical data on the other axis (usually the y-axis). 

Each category is represented by a separate bar, and the height or length of the bar corresponds to the value or frequency of the data it represents.

```python
#barplot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(x=company,height=profit)
plt.show()
```

```python
#hbarplot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(y=company,width=profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a type of data visualization that provides a summary of the distribution of numerical data. 

It displays the range, median, quartiles, and potential outliers in a concise manner.

Box plots are used to analyze and compare data across different categories or groups. They provide insights into the central tendency, spread, and 

skewness of the data, as well as any potential outliers.

The components of a box plot are as follows:

Median (Q2): It represents the middle value of the dataset when it is sorted in ascending order. It divides the data into two equal halves.

Quartiles (Q1 and Q3): These divide the data into four equal parts. Q1 represents the 25th percentile, and Q3 represents the 75th percentile. The 

interquartile range (IQR) is the difference between Q3 and Q1.

Whiskers: They represent the range of the data. The lower whisker extends to the minimum value within 1.5 times the IQR below Q1, and the upper 

whisker extends to the maximum value within 1.5 times the IQR above Q3.

Outliers: Points that lie outside the whiskers are considered outliers and are plotted individually as distinct data points.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.show()
```
